```python
## Using Keras to make predictions on pima-Indian-diabetes dataset
from keras.models import Sequential
from keras.layers import Dense
import numpy as np
```

```python
## Fix seed for reproducibility
seed = 10
np.random.seed(seed)
```

```python
## load the dataset(the csv for pima indian dataset downloaded in the same folder)
dataset = np.loadtxt("pima-indians-diabetes.csv", delimiter=",")

##split the dataset into input and target(output) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
## create model for training
model = Sequential()
model.add(Dense(9, input_dim = 8, init = 'uniform', activation = 'relu'))
model.add(Dense(8, init = 'uniform', activation = 'relu'))
model.add(Dense(1, init = 'uniform', activation = 'sigmoid'))
```

```python
## compile the model
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
## Fit the model
model.fit(X,Y, nb_epoch = 150, batch_size = 10)

scores = model.evaluate(X,Y)
print("%s : %.2f%%" %(model.metrics_names[1], scores[1]*100))
```
